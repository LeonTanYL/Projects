Creating a new 2012 dataframe containing records for Jan and Feb for concatenation afterwards.

```python
df_12_1 = df_00s[df_00s["month"].str.contains("2012")]
```

```python
df_12_1.head()
```

Using the lease commencement as reference, a lambda function was implemented to create a new column for lease remaining.

```python
df_12_1["remaining_lease"] = df_12_1["lease_commence_date"].map(lambda x: (x+99)-2012)
```

```python
df_12_1.head()
```

Rearranging the columns in the dataframe.

```python
df_12_1 = df_12_1[["month","town","flat_type","block","street_name","storey_range","floor_area_sqm","flat_model","lease_commence_date","remaining_lease","resale_price"]]
```

```python
if df_12["month"].str.contains("2012").any():
    df_12["remaining_lease"] = df_12["lease_commence_date"].map(lambda x: (x+99)-2012)
elif df_12["month"].str.contains("2013").any():
    df_12["remaining_lease"] = df_12["lease_commence_date"].map(lambda x: (x+99)-2013)
else:
    df_12["remaining_lease"] = df_12["lease_commence_date"].map(lambda x: (x+99)-2014)
```

```python
df_12.head()
```

```python
df_12 = df_12[["month","town","flat_type","block","street_name","storey_range","floor_area_sqm","flat_model","lease_commence_date","remaining_lease","resale_price"]]
```

Combining the 2012 dataframes.

```python
combined_df_12 = pd.concat([df_12_1,df_12])
```

```python
combined_df_12
```

Making sure that the 2015 dataframe's remaining lease column was in the proper format.

```python
df_15.head()
```

```python
df_17.head()
```

Stripping the rest of the words from the remaining lease column in the 2017 dataframe and converting the values to integers afterwards.

```python
df_17["remaining_lease"] = df_17["remaining_lease"].map(lambda x: x[:2])
```

```python
df_17["remaining_lease"] = df_17["remaining_lease"].astype(int)
```

```python
df_17.head()
```

Ensuring that the dataframes now had equal column lengths before concatenation.

```python
combined_df_12.columns.equals(df_15.columns)
```

```python
combined_df_12.columns.equals(df_17.columns)
```

```python
df = pd.concat([combined_df_12,df_15,df_17])
```

```python
df
```

After initial inspection, there doesn't seem to be any null values in my concatenated dataset. 

```python
df.info()
```

```python
df["flat_type"].value_counts()
```

Using a dictionary to map the number of rooms to tiers in a separate column and rearragning the order of columns afterwards.

```python
flat_tier = {"1 ROOM": 1, "2 ROOM":2, "3 ROOM":3, "4 ROOM":4, "5 ROOM": 5, "MULTI-GENERATION":6, "EXECUTIVE":7}
```

```python
df["type_tier"] = [flat_tier[i] for i in df["flat_type"]]
```

```python
df = df[["month","town","flat_type","type_tier","block","street_name","storey_range","floor_area_sqm","flat_model","lease_commence_date","remaining_lease","resale_price"]]
```

Illustrating flat type and flat model against resale price using seaborn's barplot.

```python
plt.figure(figsize=(12,8))

sns.barplot(x=df["flat_type"],y=df["resale_price"])
plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(12,8))

sns.barplot(x=df["flat_model"],y=df["resale_price"])
plt.ticklabel_format(useOffset=False, style='plain', axis='y')
plt.xticks(rotation=90);
```

Visualization of potential resale price outliers using Seaborn's boxplot.

```python
plt.figure(figsize=(12,8))

sns.boxplot(x=df["resale_price"])
plt.ticklabel_format(useOffset=False, style='plain',axis="x");
```

Changing the values in the month column to datetime format and displaying just the month and year before exporting the cleaned dataset to CSV for visualization purposes in Tableau.

```python
df["month"] = pd.to_datetime(df["month"],format="%Y-%m").dt.to_period("M")
```
